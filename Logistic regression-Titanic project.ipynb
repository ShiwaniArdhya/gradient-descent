```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df_train=pd.read_csv('0000000000002429_training_titanic_x_y_train.csv')
df_train.head
```

```python
train_data=df_train.copy()
del train_data['Cabin']
del train_data['Name']
del train_data['Ticket']
train_data.Age.fillna(train_data.Age.mean(),inplace=True)
train_data.Embarked.fillna(train_data.Embarked.mode()[0],inplace=True)
def sex1(x):
    if x=='male':
        return 1
    return 0
def embarked1(x):
    if x=='S':
        return 0
    elif x=='C':
        return 1
    else:
        return 2
train_data['Sex']=train_data.Sex.apply(sex1)
train_data['Embarked']=train_data.Embarked.apply(embarked1)
train_data
```

```python
x_train=train_data.iloc[: ,0:7].values
y_train=train_data.iloc[:,7].values
x_train.shape,y_train.shape
```

```python
df_test=pd.read_csv('0000000000002429_test_titanic_x_test.csv')
df_test.head()
```

```python
test_data=df_test.copy()
del test_data['Cabin']
del test_data['Name']
del test_data['Ticket']
test_data.Age.fillna(test_data.Age.mean(),inplace=True)
test_data.Embarked.fillna(test_data.Embarked.mode()[0],inplace=True)
test_data['Sex']=test_data.Sex.apply(sex1)
test_data['Embarked']=test_data.Embarked.apply(embarked1)
test_data
```

```python
x_test=test_data.iloc[:,:]
x_test.shape
```

```python
x_train.shape,y_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_train,x_test
```

```python
clff=LogisticRegression(solver='liblinear')
```

```python
clff.fit(x_train,y_train)
```

```python
prediction22=clff.predict(x_test)
print(prediction22)
```

```python
np.savetxt("Titanic_predictions.csv",prediction22, fmt='%d')
```
